##  data transformation  
#####  calculating avarage of last 17 years, 15 years , 12 years and so on... as required

```python
dataset.head()
```

```python
lenth=dataset.shape[0]
width=dataset.shape[1]
```

```python
dataset_year=dataset.iloc[:,1:width]

lenth_y=dataset_year.shape[0]
width_y=dataset_year.shape[1]

dataset_year.head()
```

```python
def cal_avg(n):
    # n define as: last n years of avarage
    
    avg=np.round(dataset_year.iloc[:,width_y-n:width_y].mean(axis=1),2)
    return(avg)
    
    
    
```

```python
# we have 17 years of data
dataset["avg_last_17yrs"]=cal_avg(17)
```

```python
dataset.head()
```

```python
dataset["avg_last_15yrs"]=cal_avg(15)
dataset["avg_last_12yrs"]=cal_avg(12)
dataset["avg_last_10yrs"]=cal_avg(10)
dataset["avg_last_8yrs"]=cal_avg(8)
dataset["avg_last_5yrs"]=cal_avg(5)
dataset["avg_last_3yrs"]=cal_avg(3)
dataset["avg_last_2yrs"]=cal_avg(2)
```
